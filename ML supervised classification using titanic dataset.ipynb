Applying Logistic Regression

```python
data.corr()
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder as Lb
encoder = Lb()
```

```python
data['Survived'] = encoder.fit_transform(data['Survived'])
```

```python
data['Sex'] = encoder.fit_transform(data['Sex'])
```

```python
data['Embarked'] = encoder.fit_transform(data['Embarked'])
```

```python
# feature extraction

X = data[['Parch', 'Sex', 'SibSp', 'Pclass']]
Y = data['Survived'].values
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=20)

model.fit(X_train, Y_train)
```

```python
X_test
```

```python
model.score(X_test, Y_test)
```

```python
model.score(X_train, Y_train)
```

```python
print(f'Accuracy = {model.score(X_test,Y_test)}')

print('Accuracy = {: 2f}'.format(model.score(X_test,Y_test)))
```

```python
import seaborn as sns
sns.regplot(x='Fare',y='Survived',data=data)
```

Confusion Matrix

```python
Y_pred = model.predict(X_test)

confusion_matrix = pd.crosstab(Y_test,Y_pred,rownames=['Actual'],colnames=['Predicted'])
sns.heatmap(confusion_matrix,annot = True)
```

```python
from sklearn.metrics import classification_report

print(classification_report(Y_test,Y_pred))
```

ROC

```python
from sklearn.metrics import roc_curve
#coc curve for models
fpr, tpr, thresh = roc_curve(Y_test, Y_pred, pos_label=1)

#roc curve for tpr = fpr
random_probs =[0 for i in range(len(Y_test))]
fpr, tpr , _= roc_curve(Y_test, random_probs, pos_label=1)
print(fpr, tpr, _)
```

```python
from sklearn.metrics import roc_auc_score

#auc scores
auc_score = roc_auc_score(Y_test,Y_pred)

print(auc_score)
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
#roc curve for classes
fpr = {}
tpr = {}
thresh ={}

n_class = 2

for i in range(n_class):
    fpr[i], tpr[i], thresh[i] = roc_curve(Y_test, Y_pred, pos_label=i)
    
#plottinh
plt.plot(fpr[0], tpr[0], linestyle='--',color='orange', label='Class 0 vs Rest')
plt.plot(fpr[1], tpr[1], linestyle='--',color='green', label='Class 1 vs Rest')

plt.title('Multiclass ROC curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive rate')
plt.legend(loc='best')
plt.savefig('Multiclass ROC',dpi=300);
```

KNN

```python
# extract the features
x = data.loc[:, ['Parch', 'Sex', 'SibSp', 'Pclass']]
y = data.loc[:, 'Survived']
```

```python
data.shape
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

model = KNeighborsClassifier(5)

X_train, X_test, y_train, y_test = train_test_split(x,y, test_size= 0.3, random_state=20)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print(f'The accuracy of the moel is {accuracy_score(y_pred, y_test)}')
```

```python
print(f'The accuracy of the model is {accuracy_score(model.predict(X_train), y_train)}')
```

```python
scores = []
for k in range(1,20):
  model = KNeighborsClassifier(k)
  model.fit(X_train,y_train)
  y_pred = model.predict(X_test)
  print(f'{k}:{accuracy_score(y_pred,y_test)}')
  scores.append(accuracy_score(y_pred,y_test))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (16,8))
plt.plot(range(1,20), scores)
plt.xlabel('Value K for KN')
plt.ylabel('Testing Accuracy')
plt.title('Testing Accuracy for different values of K')
```
